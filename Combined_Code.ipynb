```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
import joblib
import numpy as np
import os
from keras_preprocessing import image
import cv2
import glob
from skimage.measure.entropy import shannon_entropy


input_img = r"C:\Users\Vrushank Anand\Documents\College Related\Semester 7\Project_Ykm_Sir\Raw_Photos(train_val)\train\No object\image181.jpg"

model=Sequential()
model.add(Conv2D(16, kernel_size=(5, 5),activation='relu',input_shape=(150,150,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, kernel_size=(5, 5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
nc = 0
count = 0

folder_name = r"C:\Users\Vrushank Anand\Documents\College Related\Semester 7\Project_Ykm_Sir\Raw_Photos(train_val)\val\Object"


for filename in os.listdir(folder_name):

    # for normalized hgih pass filter
    model_path_1 = 'C:\\Users\\Vrushank Anand\\Documents\\College Related\\Semester 7\\Project_Ykm_Sir\\Codes_for_project\\object_detection_normalized_high_pass_filter_model.h5'
    model.load_weights(model_path_1)

    # test_image =image.load_img(os.path.join(folder_name,filename),target_size =(150,150))

    img = cv2.imread(os.path.join(folder_name,filename))
    gauss = cv2.GaussianBlur(img, (21,21), 3) 
    sharp_image = img - gauss
    normalized_image = np.zeros((800, 800))
    normalized_image = cv2.normalize(sharp_image, normalized_image, -127, 128, cv2.NORM_MINMAX)

    test_image =image.img_to_array(normalized_image)    
    test_image = tf.image.resize(test_image, [150,150])
    test_image =np.expand_dims(test_image, axis =0)



    result = model.predict(test_image)
    if result[0][0] >= 0.5:
        prediction_1 = 'Object'
    else:
        prediction_1 = 'No object'
    # print(prediction_1)


    # for high boost filter
    model_path_2 = r"C:\Users\Vrushank Anand\Documents\College Related\Semester 7\Project_Ykm_Sir\Codes_for_project\object_detection_high_boost_filter_model.h5"
    model.load_weights(model_path_2)

    # test_image =image.load_img(os.path.join(folder_name,filename),target_size =(150,150))


    img = cv2.imread(os.path.join(folder_name,filename))
    gauss = cv2.GaussianBlur(img, (7,7), 0) 
    sharp_image = cv2.addWeighted(img, 2, gauss, -1, 0) 

    test_image =image.img_to_array(sharp_image)
    test_image = tf.image.resize(test_image, [150,150])
    test_image =np.expand_dims(test_image, axis =0)
        
    result = model.predict(test_image)
    if result[0][0] >= 0.5:
        prediction_2 = 'Object'
    else:
        prediction_2 = 'No object'
    print(result)


    # for variance and entropy model

    test_image =image.load_img(os.path.join(folder_name,filename),target_size =(150,150))

    entropy = shannon_entropy(test_image)
    variance = np.var(test_image)

    # test_path = r"C:\Users\Vrushank Anand\Documents\College Related\Semester 7\Project_Ykm_Sir\Images_data_Test.csv"
    # test_dataset = pd.read_csv(test_path)
    test_array = [entropy,variance]
    X_validation = np.array(test_array)
    X_validation = X_validation.reshape(1,-1)
    # print(X_validation)

    lda = joblib.load(r"C:\Users\Vrushank Anand\Documents\College Related\Semester 7\Project_Ykm_Sir\Codes_for_project\Variance_Entropy_model.joblib")
    prediction_3 = lda.predict(X_validation)

    # print(prediction_3)


    if(prediction_1==prediction_2):
        main_prediction = prediction_1  
    elif(prediction_1==prediction_3):
        main_prediction = prediction_1
    elif(prediction_2==prediction_3):
        main_prediction = prediction_2
    elif(prediction_1==prediction_2==prediction_3):
        main_prediction = prediction_1
    else:
        main_prediction = 'Failure in predicting'

    # print(main_prediction)
    if(main_prediction != 'No object'):
        print(main_prediction)
        nc = nc+1
    else:
        # print("100 percent efficiency")
        count = count + 1
print(nc)
eff = count/(nc + count)
print("Efficiency:", eff)


```
